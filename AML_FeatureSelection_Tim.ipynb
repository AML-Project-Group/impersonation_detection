# Setup code

```python
import csv
import numpy
import pandas as pd
from matplotlib import pyplot
from sklearn.preprocessing import Binarizer
from numpy import set_printoptions
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import export_graphviz
import pydot
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

''' Load data and create a dataframe '''

filename = 'train_imperson_without4n7_balanced_data.csv'

df = pd.read_csv('C:\\Users\\timot\Desktop\\AMLGroupProject\\train_imperson_without4n7_balanced_data.csv', header=0, na_values=['nan'])
print(df.head())
print(df.info())
print(df.describe())

''' Delete columns with no values '''

# many features have no values
columnsToDelete = []

# if the  feature has std == 0 it means no values or only one value

for key, value in df.iteritems():
    if df[key].std() == 0:
        columnsToDelete.append(key)

df = df.drop(columnsToDelete, axis=1)
len(columnsToDelete)  # 74 features have been deleted

print(df.head())


###Using the RF

df.columns=[i for i in range(1,80)] #rename column heads

X = df.iloc[:, 0:78] #set x variables to be the first 78 columns
y = df.iloc[:,-1] #set y variable to be last column
```

# Chi2 test. We assume columns with < 6 unique values are categorical

```python
Xcat = df.loc[:, df.nunique() < 6]
print(Xcat.nunique())

columnnames = list(Xcat)
chi2,pval = chi2(Xcat,y)



feature_importances2 = [(feature,round(importance,2)) for feature,importance in zip(columnnames ,chi2)]
feature_importances2 = sorted(feature_importances2, key = lambda x: x[1], reverse = True)
[print('Variable: {:20} Chi2: {}'.format(*pair)) for pair in feature_importances2];



feature_importances2 = [(feature,round(importance,2)) for feature,importance in zip(columnnames ,pval)]
feature_importances2 = sorted(feature_importances2, key = lambda x: x[1], reverse = False)
[print('Variable: {:20} Pval: {}'.format(*pair)) for pair in feature_importances2];
```
